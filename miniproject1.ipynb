```python
from itertools import combinations

# Given set of numbers
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Function to find subsets of size 5 with sum equal to zero
def find_subsets_with_sum_zero(nums):
    subsets = []
    # Generate all combinations of size 5
    for subset in combinations(nums, 5):
        # Check if the sum of the subset is zero
        if sum(subset) == 0:
            subsets.append(subset)
    return subsets

# Find subsets with sum zero
zero_sum_subsets = find_subsets_with_sum_zero(numbers)

# Print the subsets
print("Subsets of size 5 with sum zero:")
for subset in zero_sum_subsets:
    print(subset)
```

```python
from itertools import combinations

# Given set of numbers
numbers = [-12, -3, -6, 7, 2, -2, 6, 3, 9, -7, -5, -8, 1, 11, -9, -4]

# Function to find subsets with sum equal to zero and size between 3 to 6
def find_subsets_with_sum_zero(nums):
    subsets = []
    # Iterate over subset sizes from 3 to 6
    for size in range(3, 7):
        # Generate all combinations of current size
        for subset in combinations(nums, size):
            # Check if the sum of the subset is zero
            if sum(subset) == 0:
                subsets.append(subset)
    return subsets

# Find subsets with sum zero and size between 3 to 6
zero_sum_subsets = find_subsets_with_sum_zero(numbers)

# Print the subsets
print("Subsets with sum zero and size between 3 to 6:")
for subset in zero_sum_subsets:
    print(subset)
```
